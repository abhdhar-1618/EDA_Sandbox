```python
# Create a unique list of Education levels
e_types = df.Education.unique()
ttests = []

# Loop through that list
for i, e_t in enumerate(e_types): # Use the enumerate() function to add an index for counting to a list of values
  # For each item, loop through a second list of each item to compare each pair
  for i2, e_t2 in enumerate(e_types):
    if i2 > i: # If the inner_index is greater that the outer_index, then go ahead and run a t-test
      type_1 = df[df.Education == e_t]
      type_2 = df[df.Education == e_t2]
      t, p = stats.ttest_ind(type_1['Purchased Bike'], type_2['Purchased Bike'])
      
      # Add each t-test result to a list of t, p pairs
      ttests.append([e_t + ' - ' + e_t2, t.round(4), p.round(4)])

      # Let's also print them out just to make sure it's working
      print(f'{e_t}/{e_t2}: t: {round(t, 2)}, p: {round(p, 2)}')
      
# The new threshold for statistical significant = 0.05 / length of the ttests list
p_threshold = 0.05 / len(ttests)
print(f'\nT-tests below p-value threshold: {p_threshold}')

# Now let's iterate through the results list and print only those that are significant based on the new threshold
for pair in ttests:
  if pair[2] <= p_threshold: # The p-value is in the second index, hence 1
    print(pair)
```

```python
import pandas as pd

# Assuming df is your DataFrame loaded from 'bikebuyers.csv'
df = pd.read_csv('bikebuyers.csv', usecols=['Education', 'Commute Distance', 'Purchased Bike', 'Age', 'Home Owner'])

# Create 'Education_rank' column and map values
df['Education_rank'] = df['Education']
df['Education_rank'].replace(['Partial High School', 'High School', 'Partial College', 'Bachelors', 'Graduate Degree'], 
                             [1, 2, 3, 4, 5], inplace=True)

# Create 'Commute_rank' column and map values
df['Commute_rank'] = df['Commute Distance']
df['Commute_rank'].replace(['0-1 Miles', '1-2 Miles', '2-5 Miles', '5-10 Miles', '10+ Miles'], 
                           [0, 1, 2, 5, 10], inplace=True)

# Map 'Purchased Bike' column values
df['Purchased Bike'].replace(['No', 'Yes'], [0, 1], inplace=True)

# Convert columns to int64 data type
df['Education_rank'] = df['Education_rank'].astype('int64')
df['Commute_rank'] = df['Commute_rank'].astype('int64')
df['Purchased Bike'] = df['Purchased Bike'].astype('int64')

# Display the DataFrame
df.head()
```

```python
# Generate ANOVA results
groups = df['Commute Distance'].unique()
group_labels = []
for g in groups:
  group_labels.append(df[df['Commute Distance'] == g]['Purchased Bike'])
oneway = stats.f_oneway(*group_labels)

# Begin building the text to display descriptive statistics in the diagram
textstr  = '       ANOVA' + '\n'
textstr += 'F:            ' + str(round(oneway[0], 2)) + '\n'
textstr += 'p:      ' + str(round(oneway[1], 2)) + '\n\n'
textstr += 'Sig. comparisons (Bonferroni-corrected)' + '\n'

ttests = {} # Create an empty dictionary to keep track of results

for i1, group1 in enumerate(groups):
  for i2, group2 in enumerate(groups):
    if i2 > i1:
      values1 = df[df['Commute Distance'] == group1]['Purchased Bike'] # list of 0/1 values for group_1
      values2 = df[df['Commute Distance'] == group2]['Purchased Bike'] # list of 0/1 values for group_2
      t, p = stats.ttest_ind(values1, values2)
      ttests[f'{group1} v {group2}'] = round(p, 4) # Adding the p-value only

p_threshold = 0.05 / len(ttests.values()) # Bonferroni-corrected p-value determined

# Loop through the p-values and only print those under the Bonferroni p-value
for k, v in ttests.items():
  if v <= p_threshold:
    textstr += f'{k}: p={v}\n' # Building on the textstr we started earlier with ANOVA results

# Build the plot and add the textstr
viz = sns.barplot(x="Commute Distance", y="Purchased Bike", data=df, 
                  order=['0-1 Miles', '1-2 Miles', '2-5 Miles', '5-10 Miles', '10+ Miles'],
                  capsize=.2, palette="flare");
viz.set_xticklabels(viz.get_xticklabels(), rotation=25);
plt.text(1, 0.1, textstr, fontsize=12, transform=plt.gcf().transFigure)
plt.show()
```

## Practice 3: Occupation

```python
import pandas as pd
df = pd.read_csv('bikebuyers.csv',
                 usecols=['Occupation', 'Income', 'Gender', 'Home Owner', 'Purchased Bike'])
df.head()
```

```python
# Generate ANOVA results
groups = df['Occupation'].unique()
group_labels = []
for g in groups:
  group_labels.append(df[df['Occupation'] == g]['Income'])
oneway = stats.f_oneway(*group_labels)

# Begin building the text to display descriptive statistics in the diagram
textstr  = '       ANOVA' + '\n'
textstr += 'F:            ' + str(round(oneway[0], 2)) + '\n'
textstr += 'p:      ' + str(round(oneway[1], 2)) + '\n\n'
textstr += 'Sig. comparisons (Bonferroni-corrected)' + '\n'

ttests = {} # Create an empty dictionary to keep track of results

for i1, group1 in enumerate(groups):
  for i2, group2 in enumerate(groups):
    if i2 > i1:
      values1 = df[df['Occupation'] == group1]['Income'] # list of 0/1 values for group_1
      values2 = df[df['Occupation'] == group2]['Income'] # list of 0/1 values for group_2
      t, p = stats.ttest_ind(values1, values2)
      ttests[f'{group1} v {group2}'] = round(p, 4) # Adding the p-value only

p_threshold = 0.05 / len(ttests.values()) # Bonferroni-corrected p-value determined

# Loop through the p-values and only print those under the Bonferroni p-value
for k, v in ttests.items():
  if v <= p_threshold:
    textstr += f'{k}: p={v}\n' # Building on the textstr we started earlier with ANOVA results

# Build the plot and add the textstr
viz = sns.barplot(x="Occupation", y="Income", data=df, capsize=.2, palette="flare");
viz.set_xticklabels(viz.get_xticklabels(), rotation=25);
plt.text(1, 0.1, textstr, fontsize=12, transform=plt.gcf().transFigure)
plt.show()
```

Management clearly makes the most income and it is significantly higher than all other professions.
